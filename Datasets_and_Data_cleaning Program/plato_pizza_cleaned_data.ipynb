**PLATO'S PIZZA DATA CLEANING**

**Import datasets**

There are 4 csv datasets : order_details, orders, pizza_types, pizzas


```python
import pandas as pd

order_details = pd.read_csv('order_details.csv')
orders = pd.read_csv('orders.csv')
pizza_types = pd.read_csv('pizza_types.csv', encoding='latin1')
pizzas = pd.read_csv('pizzas.csv', encoding='latin1')
```

**Conclusion**

There is no missing values
`date` and `time` attribute of table `orders` has data type `object`

`order_id` attribute of table `orders` has data type `int`

`order_id` attribute of table `order_details` has data type `int`

**Action items**

Change dtype of `date`, `time`, `order_id` in table `orders`

Change dtype of `order_id` in table `order_details`

Create an aggregated table for Power BI analysis

```python
order_details['order_id'] = order_details['order_id'].astype(str)
```

```python
orders['order_id'] = orders['order_id'].astype(str)
orders['date'] = pd.to_datetime(orders['date'], format= '%Y-%m-%d')
orders['time'] = pd.to_datetime(orders['time'], format='%H:%M:%S').dt.time
```

```python
orders_sales = pd.merge(orders, order_details, on='order_id')
orders_sales_pizzas = pd.merge(orders_sales, pizzas, on='pizza_id')
plato_pizza_aggregated = pd.merge(orders_sales_pizzas, pizza_types, on='pizza_type_id')
plato_pizza_aggregated.head()
```
